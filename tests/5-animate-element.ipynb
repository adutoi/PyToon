```python
import math
from IPython.display import SVG
from pytoon import struct, composite, line, circle, polygon, animated
```

```python
def c(_t_):
    f = 1    # 1 Hz
    x = 50 * math.cos(2*math.pi * f * _t_)
    y = 50 * math.sin(2*math.pi * f * _t_)
    return (x,y)

image = composite([
    line(begin=(-100,-100), end=(100,100), lstyle="floral white"),
    line(begin=(-100,100), end=(100,-100), lstyle="floral white"),
    circle(center=animated(c,Dt=1/100.), radius=25, line="red")
    ])

image.svg("outputs/animated-element", title="Orbit", time=(0,1), duration=5, controls=struct(scale=0.5, location=(-50,-60)))
```
