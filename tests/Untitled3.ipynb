```python
from math import pi, sin, cos
from util import int_round
from pytoon import *

def w(_t_):
    return 1.001 - sin(2*pi*_t_)

def c(_t_):
    return (0,10*w(_t_))

def g(_t_):
    c2 = cos(2*pi*_t_) ** 2
    s2 = sin(2*pi*_t_) ** 2
    rr = "{:02x}".format(int_round(c2*0xff + s2*0))
    bb = "{:02x}".format(int_round(s2*0xff + c2*0))
    return "#" + rr + "00" + bb

composite([
    line(begin=animated(c,Dt=0.05), end=(100,0), lstyle=linestyle(color=animated(g,Dt=0.05), weight=animated(w,Dt=0.05)))
]).svg(time=(0,1), duration=1)
```

```python
for i in range(101):
    t = i / 100
    print(g(t))
```
