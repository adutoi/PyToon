```python
import math
from IPython.display import SVG
from pytoon import composite, circle, polygon, positional_transform
```

```python
def compress():
    def f(p):
        x, y = p
        y = y*(2 - math.cos(2*math.pi * x / 150))
        x = x + 8*math.sin(2*math.pi * x / 150)
        return x,y
    return f, 1

positions = [i*12 for i in range(50)]
```

```python
beads = composite([
    *(polygon(points=[(b,-1), (e,-1), (e,1), (b,1)], fstyle="gray", lstyle=False) for b,e in zip(positions,positions[1:])),
    *(circle(radius=4, center=(x,0), fstyle="cyan", lstyle=("black",0.25)) for x in positions)
], transform=positional_transform(compress))

image = composite([
    #polygon(points=[(-10,20),(-10,-20),(600,-20),(600,20)], lstyle=False, fstyle="white"),
    "fig"
], fig=beads)
```

```python
image.svg("springy")
SVG("springy.svg")
```
